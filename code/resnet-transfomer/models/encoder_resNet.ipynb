```python
from torch import Tensor
import torch
import torch.nn as nn
import torchvision
from torchvision import transforms
```

```python
# in： 64，3，256，256
# out: 64,128,16,16
# out+2d pos_encode: 64,128,16,16
# flatten: 64,128,256
# 
```

```python
test_x = torch.rand(64,3,256,256)
```

```python
resnet18 = torchvision.models.resnet18(pretrained=False)
```

```python
resnet50 = torchvision.models.resnet50(pretrained=False)
```

```python
modules=list(resnet50.children())[:-4]
re_resnet50 = nn.Sequential(*modules)
bottleneck = nn.Conv2d(512, 128, 1)
```

```python
out = re_resnet50(test_x)
out = bottleneck(out)
print(out.shape)
```
